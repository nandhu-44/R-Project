```python
# Step 1: Load the data
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data files
def load_data():
    temp_annual_max = pd.read_csv("temp_dataset/TEMP_ANNUAL_MAX_1901-2021.csv")
    temp_annual_min = pd.read_csv("temp_dataset/TEMP_ANNUAL_MIN_1901-2021.csv")
    temp_annual_seasonal_mean = pd.read_csv("temp_dataset/TEMP_ANNUAL_SEASONAL_MEAN.csv")
    return temp_annual_max, temp_annual_min, temp_annual_seasonal_mean

# Load datasets
temp_annual_max, temp_annual_min, temp_annual_seasonal_mean = load_data()

# Step 2: Display basic info and first few rows
print("TEMP_ANNUAL_MAX dataset:")
print(temp_annual_max.info())
print(temp_annual_max.head())

print("\nTEMP_ANNUAL_MIN dataset:")
print(temp_annual_min.info())
print(temp_annual_min.head())

print("\nTEMP_ANNUAL_SEASONAL_MEAN dataset:")
print(temp_annual_seasonal_mean.info())
print(temp_annual_seasonal_mean.head())
```

```python
# Summary statistics for each dataset
print("Summary Statistics for TEMP_ANNUAL_MAX:")
print(temp_annual_max.describe())

print("\nSummary Statistics for TEMP_ANNUAL_MIN:")
print(temp_annual_min.describe())

print("\nSummary Statistics for TEMP_ANNUAL_SEASONAL_MEAN:")
print(temp_annual_seasonal_mean.describe())
```

```python
# Plot the annual max and min temperature trends over the years
plt.figure(figsize=(12, 6))
plt.plot(temp_annual_max['YEAR'], temp_annual_max['ANNUAL'], label='Annual Max Temperature', color='red')
plt.plot(temp_annual_min['YEAR'], temp_annual_min['ANNUAL'], label='Annual Min Temperature', color='blue')
plt.xlabel("Year")
plt.ylabel("Temperature (C)")
plt.title("Yearly Maximum and Minimum Temperature Trends (1901-2021)")
plt.legend()
plt.show()
```

```python
# Plot seasonal temperature trends for max and min temperatures
plt.figure(figsize=(12, 6))
plt.plot(temp_annual_max['YEAR'], temp_annual_max['JAN-FEB'], label='Jan-Feb Max', color='orange')
plt.plot(temp_annual_max['YEAR'], temp_annual_max['MAR-MAY'], label='Mar-May Max', color='yellow')
plt.plot(temp_annual_max['YEAR'], temp_annual_max['JUN-SEP'], label='Jun-Sep Max', color='green')
plt.plot(temp_annual_max['YEAR'], temp_annual_max['OCT-DEC'], label='Oct-Dec Max', color='purple')

plt.plot(temp_annual_min['YEAR'], temp_annual_min['JAN-FEB'], label='Jan-Feb Min', linestyle='--', color='orange')
plt.plot(temp_annual_min['YEAR'], temp_annual_min['MAR-MAY'], label='Mar-May Min', linestyle='--', color='yellow')
plt.plot(temp_annual_min['YEAR'], temp_annual_min['JUN-SEP'], label='Jun-Sep Min', linestyle='--', color='green')
plt.plot(temp_annual_min['YEAR'], temp_annual_min['OCT-DEC'], label='Oct-Dec Min', linestyle='--', color='purple')

plt.xlabel("Year")
plt.ylabel("Temperature (C)")
plt.title("Seasonal Temperature Trends (1901-2021)")
plt.legend()
plt.show()
```

```python
# Correlation analysis on the annual data
annual_data = pd.DataFrame({
    'Year': temp_annual_max['YEAR'],
    'Annual Max': temp_annual_max['ANNUAL'],
    'Annual Min': temp_annual_min['ANNUAL']
})

correlation_matrix = annual_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Between Annual Max and Min Temperatures")
plt.show()
```

```python
# Monthly variation for max and min temperatures using boxplot
months = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']

plt.figure(figsize=(14, 8))
sns.boxplot(data=temp_annual_max[months])
plt.title("Monthly Temperature Variation (Max)")
plt.xlabel("Month")
plt.ylabel("Temperature (C)")
plt.show()

plt.figure(figsize=(14, 8))
sns.boxplot(data=temp_annual_min[months])
plt.title("Monthly Temperature Variation (Min)")
plt.xlabel("Month")
plt.ylabel("Temperature (C)")
plt.show()
```

```python
# Calculate the temperature range for each year
temp_annual_range = pd.DataFrame({
    'Year': temp_annual_max['YEAR'],
    'Temperature Range': temp_annual_max['ANNUAL'] - temp_annual_min['ANNUAL']
})

# Plot the min-max temperature range fluctuation over the years
plt.figure(figsize=(12, 6))
plt.plot(temp_annual_range['Year'], temp_annual_range['Temperature Range'], color='purple', label='Annual Temperature Range')
plt.xlabel("Year")
plt.ylabel("Temperature Range (C)")
plt.title("Annual Min-Max Temperature Range Fluctuation (1901-2021)")
plt.legend()
plt.show()
```

```python
# Extract relevant columns for seasonal extremes
seasons = ['JAN-FEB', 'MAR-MAY', 'JUN-SEP', 'OCT-DEC']

# Plot seasonal extremes for maximum temperatures
plt.figure(figsize=(14, 10))

plt.subplot(2, 1, 1)
for season in seasons:
    plt.plot(temp_annual_max['YEAR'], temp_annual_max[season], label=f'{season} Max')
plt.xlabel("Year")
plt.ylabel("Max Temperature (C)")
plt.title("Seasonal Extremes in Maximum Temperatures (1901-2021)")
plt.legend()

# Plot seasonal extremes for minimum temperatures
plt.subplot(2, 1, 2)
for season in seasons:
    plt.plot(temp_annual_min['YEAR'], temp_annual_min[season], linestyle='--', label=f'{season} Min')
plt.xlabel("Year")
plt.ylabel("Min Temperature (C)")
plt.title("Seasonal Extremes in Minimum Temperatures (1901-2021)")
plt.legend()

plt.tight_layout()
plt.show()
```
